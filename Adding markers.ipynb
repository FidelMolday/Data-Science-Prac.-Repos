```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(0, 10, 20)
y_values = np.sin(x_values)

plt.plot(x_values, y_values, marker='o',label='Sine Wave')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Circle Markers')
plt.grid(True)
plt.legend()

plt.show()
```
