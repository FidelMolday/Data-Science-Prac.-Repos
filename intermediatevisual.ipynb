```python
import numpy as np
import matplotlib.pyplot as plt

num_points = 100
time = np.linspace(0,10, num_points)
frequency = 1
amplitude = 2

oscillator = amplitude * np.sin(2 * np.pi * frequency * time)

plt.plot(time,oscillator)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Simple Harmonic Oscillator')
plt.show()
```
